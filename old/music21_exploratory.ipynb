```python
from music21 import converter, note as m21note
import numpy as np

score = converter.parse('../data/hymns/Amazing_Grace_(nc)pope.mid')
print(f"Number of parts: {len(score.parts)}")
```

```python
def get_mean_pitch(part):
    pitches = [n.pitch.midi for n in part.recurse().getElementsByClass(m21note.Note)]
    return np.mean(pitches) if pitches else 0

def classify_part(mean_pitch):
    if mean_pitch > 65:
        return 'Soprano'
    elif mean_pitch > 57:
        return 'Alto'
    elif mean_pitch > 50:
        return 'Tenor'
    else:
        return 'Bass'

part_labels = {}
for i, part in enumerate(score.parts):
    mp = get_mean_pitch(part)
    label = classify_part(mp)
    part_labels[label] = part
    print(f"Track {i}: mean pitch={mp:.1f} → {label}")
```

```python
def extract_part(score, voice: str):
    """
    Extract a voice part from a hymn score.
    voice: 'Soprano', 'Alto', 'Tenor', or 'Bass'
    """
    labeled = {}
    for part in score.parts:
        mp = get_mean_pitch(part)
        label = classify_part(mp)
        labeled[label] = part

    if voice not in labeled:
        raise ValueError(f"Could not find {voice} part. Found: {list(labeled.keys())}")

    return labeled[voice]

# Example usage
soprano_part = extract_part(score, 'Soprano')
soprano_part.show('text')
```
